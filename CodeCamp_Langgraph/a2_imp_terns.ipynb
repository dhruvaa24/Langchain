```python

#State

#The state is a shared data structure that holds the current informtion or context of the entire application 

#in Simple terms , it is like the application's memory keeping track of the variable and data that nodes can access and modify as the yexecute 

#Nodes

#Nodes are indivudal function or operations that perform specific tasks within the graph 
#Each node receives input(oftent the current state), process it and produces an output or an updated state 

#Graph

#Graph in Langgraph is the overarching structure that maps out how different nodes are connected and executed .
#It visually represents the workflow, showing the sequence and conditional paths between various operations 

#Edges

#Edges are the connection between nodes that determine the flow of execution 
#They tell us which node should be executed next after the current one completes its tasks 

#Conditional Edges

# COnditional edges are specialized connections hat decide the next node to ezecute based on the specific condition or logic applied to the current state 

#Start point

# The start node is a virtually entry point in LangGraph , marking where the workflow begins 
#It doesnt perfrom any operations itslef , but serves as the designated starting position for the graphs execution
#End Point

#The End node signifies the conclusion of the workflow in Langgraph 

#Upon reaching this node, the graphs execution stops, indicating that all intended processes have been completed 

#Tools

# Tools 🛠️

# Tools are specialized functions or utilities that nodes can utilize to perform specific tasks such as fetching data from an API.

# They enhance the capabilities of nodes by providing additional functionalities.

# Nodes are part of the graph structure, while tools are functionalities used within nodes.

#ToolNodes

# ToolNode 🔎
# A ToolNode is just a special kind of node whose main job is to run a tool.

# It connects the tool’s output back into the State, so other nodes can use that information.

#StateGraph

# A StateGraph is a class in LangGraph used to buildand compile the graph structure.
# It manages the nodes, edges, and the overall state, ensuring that the workflow operates in a unified way and that data flows correctly between components.

#Runnable

# A Runnable in LangGraph is a standardized, executable component that performs a specific task within an Al workflow.

# It serves as a fundamental building block, allowing for us to create modular systems.

#Messages

# Human Message: Represents input from a user.

# System Message: Used to provide instructions or context to the model.

# Function Message: Represents the result of a function call.

# AI Message: Represents responses generated by AI models.

# Tool Message: Similar to Function Message, but specific to tool usage.
```
