```python
from dotenv import load_dotenv
import os

load_dotenv()


```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import HumanMessage
from langchain_tavily import TavilySearch
from langchain.agents import create_agent
from langchain.tools import tool
from pydantic import BaseModel, Field
```

```python
class Source(BaseModel):
    url : str = Field(description="url of the source")
    
class AgentResponse(BaseModel):
    answer : str = Field(description="answer to the user's question")
    sources : list[Source] = Field(description="sources used to answer the user's question, should be a list of urls")
```

```python
llm = ChatOpenAI(model="openai/gpt-4o-mini", temperature=0)
tools = [TavilySearch()]
agent = create_agent(llm, tools, response_format=AgentResponse)
```

```python
result = agent.invoke({"messages": [HumanMessage(content="Give me 5 jobs in the field of langchain")]})
```

```python
print(result["structured_response"].answer)
```

```python
print(result["structured_response"].sources)
```
