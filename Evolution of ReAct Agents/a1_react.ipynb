```python
from dotenv import load_dotenv
import os

load_dotenv()


```

```python
from langchain_classic import hub
```

```python
from langchain_classic.agents import AgentExecutor
from langchain_classic.agents.react.agent import create_react_agent
```

```python
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch
```

```python
tools = [TavilySearch()]
llm = ChatOpenAI(model="openai/gpt-4o-mini")
react_prompt = hub.pull('hwchase17/react')
```

```python
print(react_prompt.pretty_print())
```

```python
agent = create_react_agent(llm, tools, react_prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
```

```python
chain = agent_executor
result = chain.invoke(input = {"input": "Search for 3 job postings for an ai engineer using langchain in the bay area on linkedin and  list thier details."})
```
